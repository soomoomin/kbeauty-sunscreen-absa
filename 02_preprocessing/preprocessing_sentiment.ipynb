### 리뷰 전처리 및 감성분류 (KR / GB)

## preprocessing 개요

크롤링으로 수집한 국내(KR)·글로벌(GB) 리뷰 원문에 **텍스트 정제(전처리)** 를 적용하고, 사전학습된 감성분석 모델로 **긍정/부정 라벨**을 부여하는 과정을 정리한 것입니다.

### 처리 순서
1. **텍스트 정제** — 이모지·특수문자 제거, 줄바꿈 처리, 중복 공백 정리, 빈 텍스트 제거
2. **감성분류** — 사전학습 모델로 리뷰 텍스트에 긍정/부정 라벨과 확신도(score) 부여

### 사용 모델
| 대상 | 모델 | 비고 |
|---|---|---|
| 국내(KR) | `Copycats/koelectra-base-v3-generalized-sentiment-analysis` | 한국어 특화 |
| 글로벌(GB) | `nlptown/bert-base-multilingual-uncased-sentiment` | 다국어(별점 1~5 기반) |

### ⚠️ 참고
- 리뷰 원문이 담긴 CSV 파일 경로는 실행 환경에 맞게 직접 수정해야 합니다. (본 저장소에는 원본 리뷰 데이터를 포함하지 않습니다.)
- 국내(KR) 감성분류는 로컬(Jupyter) 환경, 글로벌(GB) 감성분류는 GPU 자원이 필요해 Google Colab 환경에서 진행했습니다.


## 1. 국내(KR) 리뷰 전처리

```python
import pandas as pd
import re

# 크롤링 단계에서 저장한 CSV 경로 (실행 환경에 맞게 수정)
kr = pd.read_csv('kr_combined.csv')
gb = pd.read_csv('gb_combined.csv')

print(f'국내: {len(kr)}건 / 글로벌: {len(gb)}건')
```

```python
# 전처리 함수 정의

def clean_korean(text):
    if pd.isna(text): return ''
    text = str(text)
    # 이모지 제거
    text = re.sub(r'[^\w\s가-힣ㄱ-ㅎㅏ-ㅣ]', ' ', text)
    # 줄바꿈 처리
    text = text.replace('\\r\\n', ' ').replace('\\n', ' ')
    # 중복 공백 제거
    text = re.sub(r'\s+', ' ', text).strip()
    return text

def clean_english(text):
    if pd.isna(text): return ''
    text = str(text).lower()
    # 영문+공백만 남기기
    text = re.sub(r'[^a-z\s]', ' ', text)
    # 중복 공백 제거
    text = re.sub(r'\s+', ' ', text).strip()
    return text
```

```python
# 전처리 적용
kr['review_clean'] = kr['review_txt'].apply(clean_korean)
gb['review_clean'] = gb['review_txt'].apply(clean_english)
```

```python
# 빈 텍스트 제거
print(f'드랍 전 — 국내: {len(kr)}건 / 글로벌: {len(gb)}건')

kr = kr[kr['review_clean'] != ''].reset_index(drop=True)
gb = gb[gb['review_clean'] != ''].reset_index(drop=True)

print(f'드랍 후 — 국내: {len(kr)}건 / 글로벌: {len(gb)}건')
```

```python
# 전처리 결과 저장 (경로는 실행 환경에 맞게 수정)
kr.to_csv('kr_reviews_clean.csv', index=False, encoding='utf-8-sig')
gb.to_csv('gb_reviews_clean.csv', index=False, encoding='utf-8-sig')
print('저장 완료')
```

```python
# 커널을 재시작했거나 이 지점부터 이어서 실행할 경우를 대비해,
# 정제된 파일을 다시 불러옵니다 (직전 셀 실행 후 바로 이어간다면 생략 가능)
kr = pd.read_csv('kr_reviews_clean.csv')
gb = pd.read_csv('gb_reviews_clean.csv')
print(f'국내: {len(kr)}건 / 글로벌: {len(gb)}건')
```

## 2. 국내(KR) 감성분류 — KoELECTRA

```python
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline

# 모델 로드
tokenizer = AutoTokenizer.from_pretrained(
    'Copycats/koelectra-base-v3-generalized-sentiment-analysis'
)
model = AutoModelForSequenceClassification.from_pretrained(
    'Copycats/koelectra-base-v3-generalized-sentiment-analysis'
)

# top_k=None: 긍정/부정 두 클래스의 확률을 모두 반환
clf_kr = pipeline(
    'sentiment-analysis',
    model=model,
    tokenizer=tokenizer,
    top_k=None,
    device=0 if torch.cuda.is_available() else -1
)
```

```python
# 상품 코드 → 상품명 매핑 (예시. 실제 프로젝트에서는 10개 상품의 goods_no를 매핑하여 사용)
product_map_kr = {
    'A000000000001': '1_product_a',
    'A000000000002': '2_product_b',
    # ... 나머지 상품 매핑 생략
}

kr['product_name'] = kr['goods_no'].map(product_map_kr)
```

```python
def get_sentiment(scores):
    """
    확신도 기준 없이 긍정/부정만 분류
    긍정(1) 확률이 높으면 긍정, 부정(0) 확률이 높으면 부정
    """
    pos_score = next(r['score'] for r in scores if r['label'] == '1')
    neg_score = next(r['score'] for r in scores if r['label'] == '0')

    if pos_score >= neg_score:
        return '긍정', pos_score
    else:
        return '부정', neg_score
```

```python
# 전체 리뷰에 감성 라벨 부여 (배치 처리)
texts = kr['review_clean'].tolist()
results_all = []
batch_size = 32

for i in range(0, len(texts), batch_size):
    batch = texts[i:i + batch_size]
    preds = clf_kr(batch, truncation=True, max_length=128)
    results_all.extend(preds)

kr[['sentiment_label', 'sentiment_score']] = [get_sentiment(r) for r in results_all]

print(kr['sentiment_label'].value_counts())
```

```python
# 감성분류 결과 저장
kr.to_csv('kr_sentiment.csv', index=False, encoding='utf-8-sig')
print('저장 완료')
```

## 3. 글로벌(GB) 감성분류 — nlptown/bert-base-multilingual-uncased-sentiment

```python
# ⚠️ 글로벌(GB) 감성분류는 GPU 자원이 필요해 별도로 Google Colab 환경에서 진행했습니다.
# Colab에서 새 세션으로 실행할 경우, 정제된 gb 파일을 다시 업로드해야 합니다.
#
# from google.colab import files
# uploaded = files.upload()  # gb_reviews_clean.csv 업로드

gb = pd.read_csv('gb_reviews_clean.csv')
print(f'글로벌: {len(gb)}건')
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline

# 모델 로드 (별점 1~5 기반 다국어 감성분석 모델)
tokenizer_gb = AutoTokenizer.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
model_gb = AutoModelForSequenceClassification.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')

clf_gb = pipeline(
    'sentiment-analysis',
    model=model_gb,
    tokenizer=tokenizer_gb,
    top_k=None,
    device=0 if torch.cuda.is_available() else -1
)
```

```python
# 상품 코드 → 상품명 매핑 (예시. 실제 프로젝트에서는 10개 상품의 prdtNo를 매핑하여 사용)
product_map_gb = {
    'GA000000001': '1_product_a',
    'GA000000002': '2_product_b',
    # ... 나머지 상품 매핑 생략
}

gb['product_name'] = gb['goods_no'].map(product_map_gb)
```

```python
def get_sentiment_nlptown(scores):
    """
    nlptown 라벨: '1 star'~'5 stars'
    1~2 → 부정, 4~5 → 긍정, 3 → 중립(긍정/부정 중 높은 쪽으로 병합)
    """
    score_map = {r['label']: r['score'] for r in scores}
    pos = score_map.get('4 stars', 0) + score_map.get('5 stars', 0)
    neg = score_map.get('1 star', 0) + score_map.get('2 stars', 0)

    if pos >= neg:
        return '긍정', pos
    else:
        return '부정', neg
```

```python
# 전체 리뷰에 감성 라벨 부여 (배치 처리)
texts = gb['review_clean'].tolist()
results_all = []
batch_size = 64

for i in range(0, len(texts), batch_size):
    batch = texts[i:i + batch_size]
    preds = clf_gb(batch, truncation=True, max_length=128)
    results_all.extend(preds)

gb[['sentiment_label', 'sentiment_score']] = [get_sentiment_nlptown(r) for r in results_all]

print(gb['sentiment_label'].value_counts())
```

```python
# 감성분류 결과 저장
gb.to_csv('gb_sentiment.csv', index=False, encoding='utf-8-sig')
print('저장 완료')
```
